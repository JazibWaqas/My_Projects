```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
import torch
import pandas as pd

from google.colab import drive

drive.mount('/content/drive')

# Load the saved model and tokenizer
load_path = "/content/drive/MyDrive/FinalTunedModel"
tokenizer = MBart50TokenizerFast.from_pretrained(load_path)
model = MBartForConditionalGeneration.from_pretrained(load_path)

# Set source and target languages (Urdu in your case)
tokenizer.src_lang = "ur_PK"
tokenizer.tgt_lang = "ur_PK"

# Move model to GPU
model.to("cuda")

# Define the function to generate summaries
def generate_summary(text):
    inputs = tokenizer(text, return_tensors="pt", max_length=512, truncation=True).to("cuda")
    summary_ids = model.generate(
        inputs["input_ids"],
        max_length=512,  # Shorter summaries for coherence
        num_beams=5,  # More beams for diversity
        no_repeat_ngram_size=3,  # Avoid 3-gram repetition
        length_penalty=1.2,  # Penalize overly long outputs
        early_stopping=True
    )
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Example direct input text (instead of reading from Excel)
input_texts = [
    "گلوبل وارمنگ، جسے عالمی درجہ حرارت میں اضافہ بھی کہا جاتا ہے، دنیا کے سب سے اہم اور تشویشناک ماحولیاتی مسائل میں سے ایک ہے جو زمین کے اوسط درجہ حرارت میں مسلسل اضافے کی صورت میں ظاہر ہو رہا ہے۔ یہ مسئلہ بنیادی طور پر انسانی سرگرمیوں کی وجہ سے پیدا ہوا ہے، جن میں فوسل ایندھن کا بے تحاشہ استعمال، جنگلات کی کٹائی، صنعتی سرگرمیاں، اور زراعت کے ناقص طریقے شامل ہیں، جو گرین ہاؤس گیسز کے اخراج میں اضافے کا باعث بنتے ہیں۔ گرین ہاؤس گیسز، جیسے کاربن ڈائی آکسائیڈ، میتھین، نائٹرس آکسائیڈ، اور کلورو فلورو کاربنز، زمین کے ماحول میں حرارت کو پھنساتی ہیں، جس سے \"گرین ہاؤس ایفیکٹ\" پیدا ہوتا ہے اور زمین کا درجہ حرارت بڑھنے لگتا ہے۔ اس اضافے کے نتیجے میں نہ صرف دنیا کے مختلف خطوں میں موسمیاتی نظام میں تبدیلیاں رونما ہو رہی ہیں بلکہ یہ انسانی زندگی، حیوانات، نباتات، اور قدرتی وسائل پر بھی گہرے اور خطرناک اثرات مرتب کر رہا ہے۔ گلوبل وارمنگ کے اثرات قطب شمالی اور قطب جنوبی کے گلیشیئرز کے پگھلنے سے لے کر سمندری سطح کے بلند ہونے تک ہر جگہ دیکھے جا سکتے ہیں۔ گلیشیئرز کا تیزی سے پگھلنا نہ صرف سمندری سطح کو بلند کر رہا ہے بلکہ ساحلی علاقوں کو بھی شدید خطرات سے دوچار کر رہا ہے، کیونکہ کئی ساحلی شہر اور بستیاں سمندر میں ڈوبنے کے خطرے سے قریب آ گئی ہیں۔ سمندری سطح کے بلند ہونے کی وجہ سے نمکین پانی زمینی پانی کے ذخائر میں شامل ہو رہا ہے، جس سے کھیتی باڑی اور پینے کے پانی کی قلت کا مسئلہ شدت اختیار کر رہا ہے۔ مزید برآں، موسمیاتی تبدیلیاں دنیا کے مختلف حصوں میں شدید موسموں کا باعث بن رہی ہیں، جیسے طوفان، قحط، ہیٹ ویوز، اور غیر معمولی بارشیں، جن کی شدت اور تعداد میں اضافہ ہو رہا ہے۔ زراعت پر گلوبل وارمنگ کے اثرات انتہائی سنگین ہیں، کیونکہ زمین کے درجہ حرارت میں اضافے سے فصلوں کی پیداوار میں کمی آ رہی ہے اور زراعت کے لیے درکار پانی کی دستیابی محدود ہو رہی ہے۔ گرم موسم میں کچھ فصلیں مکمل طور پر ناکارہ ہو جاتی ہیں، اور ان کی جگہ لینے والی فصلیں کم پیداوار دیتی ہیں، جس سے خوراک کی کمی کا خطرہ بڑھ رہا ہے۔ خوراک کی قیمتوں میں اضافہ اور قلت کا سامنا خاص طور پر ترقی پذیر ممالک میں غربت اور بھوک کی شدت کو بڑھا رہا ہے۔ انسانی صحت پر بھی گلوبل وارمنگ کے اثرات نمایاں ہیں، کیونکہ درجہ حرارت میں اضافے سے گرمی سے پیدا ہونے والی بیماریاں، جیسے ہیٹ اسٹروک، ملیریا، اور ڈینگی، زیادہ عام ہو رہی ہیں۔ مزید برآں، ہوا اور پانی کے معیار میں کمی انسانی صحت کو مزید خطرات سے دوچار کر رہی ہے۔ حیاتیاتی تنوع پر گلوبل وارمنگ کے اثرات انتہائی تباہ کن ہیں، کیونکہ کئی جانوروں اور پودوں کی انواع اپنی رہائش گاہوں سے محروم ہو رہی ہیں اور بعض اقسام معدومی کے خطرے سے دوچار ہیں۔ جنگلات، جو زمین کے ماحول کو متوازن رکھنے میں اہم کردار ادا کرتے ہیں، گلوبل وارمنگ کی وجہ سے متاثر ہو رہے ہیں، کیونکہ درختوں کی شرح نمو کم ہو رہی ہے اور کئی علاقوں میں جنگلات ختم ہو رہے ہیں۔ یہ مسئلہ نہ صرف کاربن کے قدرتی ذخائر کو کمزور کر رہا ہے بلکہ قدرتی ماحولیاتی نظام کے توازن کو بھی بگاڑ رہا ہے۔ گلوبل وارمنگ سے نمٹنے کے لیے فوری اور مشترکہ اقدامات کی ضرورت ہے۔ سب سے پہلے، گرین ہاؤس گیسز کے اخراج کو کم کرنا لازمی ہے، جس کے لیے قابل تجدید توانائی کے ذرائع، جیسے شمسی، ہوائی، اور آبی توانائی، کو فروغ دینا ہوگا۔ فوسل ایندھن کے استعمال کو کم کرنے، توانائی کی بچت کرنے، اور زیادہ ماحول دوست ٹیکنالوجیز کو اپنانے کی ضرورت ہے۔ جنگلات کی بحالی اور درخت لگانے کے اقدامات گلوبل وارمنگ کے اثرات کو کم کرنے میں اہم کردار ادا کر سکتے ہیں، کیونکہ درخت کاربن ڈائی آکسائیڈ کو جذب کرنے میں مدد کرتے ہیں۔ بین الاقوامی معاہدے، جیسے پیرس معاہدہ، اس مسئلے کو عالمی سطح پر حل کرنے کے لیے اہم ہیں، لیکن ان معاہدوں کے موثر نفاذ کے لیے تمام ممالک کو اپنے عزم پر قائم رہنا ہوگا اور مزید عملی اقدامات کرنے ہوں گے۔ عوامی سطح پر شعور بیدار کرنا بھی انتہائی ضروری ہے، کیونکہ ہر فرد کے چھوٹے اقدامات، جیسے پلاسٹک کے استعمال کو کم کرنا، توانائی کی بچت کرنا، اور ماحول دوست طرز زندگی اپنانا، اجتماعی طور پر بڑے اثرات پیدا کر سکتے ہیں۔ تعلیمی اداروں، میڈیا، اور حکومتی پروگرامز کے ذریعے عوام کو گلوبل وارمنگ کے خطرات اور ان سے نمٹنے کے طریقوں کے بارے میں آگاہ کیا جا سکتا ہے۔ گلوبل وارمنگ کا مسئلہ اتنا بڑا اور پیچیدہ ہے کہ اس کے حل کے لیے حکومتوں، ماہرین، تنظیموں، اور عام افراد سب کو مل کر کام کرنا ہوگا۔ یہ نہ صرف موجودہ نسلوں کی ذمہ داری ہے بلکہ آنے والی نسلوں کے لیے زمین کو محفوظ اور قابل رہائش بنانا ہمارا اخلاقی فرض بھی ہے۔ اگر فوری اقدامات نہ کیے گئے تو گلوبل وارمنگ کے اثرات مزید سنگین ہوں گے، اور یہ مسئلہ دنیا کے لیے ایک ناقابل حل چیلنج بن جائے گا۔"
]

# Generate and print summaries for each input text
for text in input_texts:
    generated_summary = generate_summary(text)

    # Print the original text and generated summary
    print("Original Text:", text)
    print("Generated Summary:", generated_summary)
    print("="*50)  # Separator for better readability

```
